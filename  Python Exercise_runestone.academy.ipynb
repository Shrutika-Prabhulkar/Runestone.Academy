Many people keep time using a 24 hour clock (11 is 11am and 23 is 11pm, 0 is midnight). If it is currently 13 and you set your alarm to go off in 50 hours, it will be 15 (3pm). Write a Python program to solve the general version of the above problem. 

Ask the user for the time now (in hours), and then ask for the number of hours to wait for the alarm. Your program should output what the time will be on the clock when the alarm goes off.

```python
time_now = input("Enter the current time ")
time_now  = int(time_now)

alarm_hour = input("Enter the alarm hour ")
alarm_hour = int(alarm_hour)

hours = time_now + alarm_hour

time_of_day = hours %24
time_of_day
```

Write a program that will convert degrees celsius to degrees fahrenheit.

```python
celsius = input("Enter celsius ")
celsius = float(celsius)

fahrenheit = (celsius * 9/5) + 32 

print("fahrenheit ",fahrenheit,"F")
```

Write a Python program that assigns the principal amount of 10000 to variable P, assign to n the value 12, and assign to r the interest rate of 8% (0.08). Then have the program prompt the user for the number of years, t, that the money will be compounded for. Calculate and print the final amount after t years.

```python
p= 10000
n =12
r = 0.08
t = input("number of years ")
t= int(t)

final_amount = p*((1+ r/n)**(n*t))
final_amount
```

Write a general version of the program which asks for the starting day number, and the length of your stay, and it will tell you the number of day of the week you will return on.

```python
vacation_start = int(input("Enter day number "))

length_vacation = int(input("Enter number of days of vacation "))

day_of_week = (length_vacation %7) + vacation_start
day_of_week
```
